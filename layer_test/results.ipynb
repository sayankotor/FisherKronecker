```python
import json
from pprint import pprint
import matplotlib.pyplot as plt
import numpy as np
```

## Per layer

```python
with open('resultsone.json', 'r') as fp:
    d = json.load(fp)
```

```python
d.keys()
```

```python
svd_l_2 = [d['svd']['2'], d['svd']['3'], d['svd']['4']]
kron_l_2 = [d['kron']['2'], d['kron']['3'], d['kron']['4']]
fwswd_l_2 = [d['fwsvd']['2'],d['fwsvd']['3'],d['fwsvd']['4']]
```

```python
ranks_ratio = [0.01, 0.05, 0.1, 0.5]
```

```python
fig = plt.figure(figsize=(36, 6))
for i in range(1, 4):
    ax = plt.subplot(1, 4, i)
    # plot lines
    ax.plot(ranks_ratio, svd_l_2[i-1], label = "SVD")
    ax.plot(ranks_ratio, kron_l_2[i-1], label = "Kron")
    ax.plot(ranks_ratio, fwswd_l_2[i-1], label = "FwSVD")
    ax.legend()
    ax.set_xlabel('rank ratio',fontsize=16)
    ax.set_ylabel('score',fontsize=16)
    ax.set_title("layer " + str(i+1),fontsize=16)
    
plt.legend()
plt.show()
```
